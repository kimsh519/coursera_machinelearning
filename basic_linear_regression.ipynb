```python
# 두 개의 데이터만을 사용한 선형 회귀

import numpy as np
import matplotlib.pyplot as plt # 라이브러리 불러오기기
```

```python
# 1. 좌표 생성
x_train = np.array([1.0,2.0])
y_train = np.array([400.0,800.0])
print('x_train = {}'.format(x_train))
print('y_train = {}'.format(y_train)) # 두 개의 x,y 좌표 생성
```

```python
# 2. x_i , y_i
for i in range(2) :
  x_i = x_train[i]
  y_i = y_train[i]
  print('x^(i) : {}, y^(i) : {} '.format(x_i,y_i)) 
  # i 번째
```

```python
# 3.데이터를 좌표상에 표현 (plot)
plt.scatter(x_train,y_train) #점 찍기
plt.title('Housing prices') #표 제목 설정
plt.ylabel('Price')
plt.xlabel('Size') #x축, y축 이름 설정
plt.show()
```

```python
# 4. 기울기 (w), y 절편 (b) 으로 직선의 방정식을 구하고 y값 도출
w = 100
b= 100 # w,b 를 100으로 가정정
def compute_y(x,w,b) :
  m = x.shape[0] # 0번째 열의 행 수 ?
  f_wb = np.zeros(m) # 행렬을 초기화
  for i in range(m) :
    f_wb[i] = w*x[i]+b # (f_wb 값 구하기)
  return f_wb

```

```python
y_value = compute_y(x_train,w,b)
plt.plot(x_train,y_value,label='prediction')
plt.scatter(x_train,y_train, label = 'data') #점 찍기
plt.title('Housing prices') #표 제목 설정
plt.ylabel('Price')
plt.xlabel('Size') #x축, y축 이름 설정
plt.legend() #범례 표시
plt.show()              #데이터와 line을 비교
```
